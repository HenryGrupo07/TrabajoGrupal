```python
def to_num(number):
    if type(number) == str:
        num = number.replace(',','')
        numb = int(num)
    else:
        numb = number
    return numb
```

```python
col = ['Population', 'Violent crime','Murder and nonnegligent manslaughter', 'Rape (revised definition)1', 'Robbery', 'Aggravated assault','Property crime', 'Burglary', 'Larceny-theft', 'Motor vehicle theft','Arson3']
for i in col:
    df[i] = df[i].apply(lambda x: to_num(x))
```
