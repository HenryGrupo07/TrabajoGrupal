Vamos a analizar que servicios tienen mas "peso" es decir: los servicios que tengan mayor suma en valores absolutos y aquellos que multiplicados por el rating promedio tengan mayores valors, para analizar si siguen la misma tendencia

```python
def ordenar_columnas_por_suma(dataframe, columnas_no_sumar=None):
    """
    Función que recibe un dataframe con valores binarios y devuelve una serie con la suma de cada columna binaria con el nombre de esa columna, ordenadas de mayor a menor, ignorando las columnas que no se quieren sumar.
    """
    if columnas_no_sumar is None:
        columnas_no_sumar = []
    columnas_a_sumar = [c for c in dataframe.columns if c not in columnas_no_sumar and dataframe[c].dtype == 'float64']
    suma_por_columna = dataframe[columnas_a_sumar].sum().sort_values(ascending=False)
    return suma_por_columna
```

```python
## Aqui sacamos las columnas que no queremos que la funcion sume
ordenar_columnas_por_suma(df, columnas_no_sumar=['name_hotel','address', 'gmap_id', 'description', 'latitude', 'longitude', 'category', 'avg_rating','num_of_reviews','price', 'hours','MISC', 'state', 'relative_results', 'etl_timestamp', 'state_xx', 'state_name', 'city_name', 'coordenadas', 'reverse_coor', 'cat_name'])
```

```python
## ahora vemos el rating promedio multiplicado por cada columna de servicio

import pandas as pd

def multiply_and_sum(df, column_name):
    # Seleccionar sólo las columnas numéricas a partir del índice 21
    cols = df.iloc[:, 21:].select_dtypes(include=['float']).columns
    
    # Multiplicar cada columna por la columna especificada
    multiplied = df[cols].multiply(df[column_name], axis=0)
    
    # Sumar los resultados por columna
    summed = multiplied.sum(axis=0)
    
    # Ordenar los resultados de mayor a menor
    sorted_results = summed.sort_values(ascending=False)
    
    return sorted_results
```
